## Detrending using linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some example data
x = np.arange(0, 10, 0.1)
y = 2*x + 5 + np.random.normal(0, 1, len(x))

plt.plot(x,y); plt.show()
```

```python
# Fit a linear regression model
coefs = np.polyfit(x, y, 1)
trend = np.polyval(coefs, x)

plt.plot(x, trend) ; plt.show()
```

```python
# Detrend the data
detrended = y - trend

plt.plot(x,detrended); plt.show()
```

```python
# Plot the original data and the detrended data
plt.plot(x, y, label='Original')
plt.plot(x, trend, label='Trend')
plt.plot(x, detrended, label='Detrended')
plt.legend()
plt.show()
```

## Detrend using differencing

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate some example data
x = np.arange(0, 10, 0.1)
y = np.sin(x) + np.random.normal(0, 0.1, len(x))

plt.plot(x,y) ; plt.show()
```

```python
# Take the first difference of the data
diff = np.diff(y)

plt.plot(x[:-1], diff) ; plt.show()
```

```python
# Plot the original data and the detrended data
plt.plot(x[:-1], y[:-1], label='Original')
plt.plot(x[:-1], diff, label='Detrended')
plt.legend()
plt.show()
```
